```
import pandas as pd
import chess
import numpy as np
from datetime import datetime, timedelta

def getPiecePositions(piece_type, piece_color):
    """
      Find all the pieces of given type and their positions
    """
    positions = []

    for square in chess.SQUARES:
        piece = board.piece_at(square)
        if piece is not None and piece.piece_type == piece_type and piece.color == piece_color:
          # In python-chess, files are numbered from 0 to 7 (where 0 corresponds to the 'a' file and 7 corresponds to the 'h' file),
          # and ranks also range from 0 to 7 (with 0 being rank 1 and 7 being rank 8).
          # You can adjust these values if you need them to match standard chess notation.
          x = int(chess.square_file(square))
          y = int(chess.square_rank(square))
          positions.append({"coord":[x, y], "square":chess.square_name(square)})  # Store the position in algebraic notation

    return positions

# Load CSV file into a DataFrame
df = pd.read_csv(filtered_output_file)
df['Queen_Pawn'] = False
df['Queen_Bishop'] = False
df['puzzle_ref'] = 'https://lichess.org/training/' + df['PuzzleId']
df['Mating_Color'] = "None"
df['square_0_0'] = "None"

# Process each row using itertuples
line=0
matched_lines=0
start_time = datetime.now()
for row in df.itertuples(index=True):

    if (line % 100000) == 0:
      # Calculate the time delta
      time_delta = datetime.now() - start_time
      print(f"{time_delta}: puzzles searched: {line} matches: {matched_lines}")

    #if df['PuzzleId'][line] != '3vD0E':
    #  line = line + 1
    #  continue

    moves_to_apply = df['Moves'][line].split()

    # Print the board position
    #print(board)
    #display(board)
    #print('Making moves')
    # Load the position from the FEN string
    board = chess.Board(df['FEN'][line])

    # Apply each move to the board
    for move in moves_to_apply:
        # Create a Move object from the UCI string
        chess_move = chess.Move.from_uci(move)

        # Check if the move is legal before making it
        if chess_move in board.legal_moves:
            board.push(chess_move)
        else:
            print(f"Illegal move: {move}")

    mated_color = None
    mating_color = None
    # Check if the position is checkmate
    if board.is_checkmate():
        # Determine the color of the king being mated
        # If it's white's turn to move, then black's king is in checkmate; otherwise, white's king is in checkmate.
        if board.turn == chess.WHITE:
            mated_color = chess.WHITE
            mating_color = chess.BLACK
        else:
            mated_color = chess.BLACK
            mating_color = chess.WHITE
    else:
        line = line + 1        # not looking at a check mate so lets ignore this
        continue

    # find pieces that deliver the check
    # Determine the square of the king in check

    # Now let's find all the queens and their positions
    king = getPiecePositions(chess.KING, mated_color)
    queens = getPiecePositions(chess.QUEEN, mating_color)

    king_pos = np.array(king[0]["coord"])

    valid_queens = []             # Find a valid distance between king and queen
    valid_pattern = False
    valid_pattern_pawn = False
    valid_pattern_bishop = False
    for queen in queens:
      queen_pos = np.array(queen["coord"])
      delta_queen = king_pos - queen_pos
      # Check if queen is in right/left and above/below square of king
      if abs(delta_queen[1])  == 1 and abs(delta_queen[0]) == 1:
        valid_queens.append(queen)

    if len(valid_queens) == 0:
      line = line + 1
      continue

    # So, we have found a restricting  Queen, find for pawn defending the Queen
    valid_pawns = []
    pawns = getPiecePositions(chess.PAWN, mating_color)
    for queen in valid_queens:
      if valid_pattern is True:      # Exit here if pattern is already confirmed
        break

      queen_pos = np.array(queen["coord"])
      delta_queen = king_pos - queen_pos

      for pawn in pawns:
        pawn_pos = np.array(pawn["coord"])
        delta_pawn = queen_pos - pawn_pos

        if df['PuzzleId'][line] == '7F6Je':   # DEBUG
          print(f"DEBUG Row {line} {df['PuzzleId'][line]}: queen_pos: {queen_pos}, pawn_pos:{pawn_pos} ")

        # Check if pawn is in right/left and above/below square of queen
        # and if queen and pawn are on one diagonal line defending
        # with pawn defending queen against the king
        if mating_color == chess.WHITE:
          if abs(delta_pawn[0])  == 1 and delta_pawn[1] == 1:
            if df['PuzzleId'][line] == '7F6Je':   # DEBUG
              print(f"DEBUG Row {line} {df['PuzzleId'][line]}: valid mating WHITE delta_pawn:{delta_pawn} ")

            valid_pattern = True
            valid_pattern_pawn = True
            break
        else:
          if abs(delta_pawn[0])  == 1 and delta_pawn[1] == -1:
            if df['PuzzleId'][line] == '7F6Je':   # DEBUG
              print(f"DEBUG Row {line} {df['PuzzleId'][line]}: valid mating BLACK delta_pawn:{delta_pawn} ")

            valid_pattern = True
            valid_pattern_pawn = True
            break

      # Check if we can find a bishop instead of a pawn defending the Queen
      if valid_pattern is False:
        bishops = getPiecePositions(chess.BISHOP, mating_color)
        for bishop in bishops:
          bishop_pos = np.array(bishop["coord"])
          delta_bishop = queen_pos - bishop_pos

          # Check if bishop is in right/left and above/below square of queen
          # and if queen and bishop are on one diagonal line defending each other
          # against the king
          # see if bishop is on a diagonal to the king
          if abs(delta_bishop[0]) > 0 and abs(delta_bishop[0])  == abs(delta_bishop[1]):

            # Now lets make sure there is no other piece between bishop and queen
            valid_diagonal = True

            xstart = queen_pos[0]           # Adjust x and y range
            xend   = bishop_pos[0]         # such that we can loop from start to end
            if xstart > xend:
              xstart = bishop_pos[0]
              xend   = queen_pos[0]

            ystart = queen_pos[1]
            yend   = bishop_pos[1]
            if ystart > yend:
              ystart = bishop_pos[1]
              yend   = queen_pos[1]

            #print(f"PuzzleId: {df['PuzzleId'][line]} - testing x:({xstart}, {xend}), y:({ystart}, {yend})")
            y = ystart
            for x in range(xstart, xend):
              square=chess.square(file_index=x, rank_index=y)
              piece = board.piece_at(square)

              #print(f"{piece} at {x}, {y}")

              if piece is not None:
                if piece.piece_type == chess.BISHOP:
                  if piece.color == mated_color:
                    valid_diagonal = False
                    break

                if piece.piece_type == chess.QUEEN:
                  if piece.color == mating_color:
                    valid_diagonal = False
                    break

                if piece.piece_type != chess.BISHOP and piece.piece_type != chess.QUEEN:
                  valid_diagonal = False
                  break

              y = y + 1   # Keep searching the diagonal line betweeb king and bishop

            if valid_diagonal:
              valid_pattern = True
              valid_pattern_bishop = True
              break

    # Print the board position
    #print(board)
    #print(f"Row {line} {df['PuzzleId'][line]} | {df['FEN'][line]} | {moves_to_apply}| | {df['GameUrl'][line]}")
    #print(f"Positions of {'white' if mated_color else 'black'}'s king {king} {'white' if mating_color else 'black'} queens {queens} and rooks {rooks}")
    #print(f"{'white' if mated_color else 'black'}'s king is checkmated by {'white' if mating_color else 'black'}.")
    #print(f"Positions of {'white' if mated_color else 'black'}'s king {king} mated by {'white' if mating_color else 'black'} queens {mating_queens} and rooks {mating_rooks}")
    #print(f"Mated by {'white' if mating_color else 'black'} queens {mating_queens} and rooks {mating_rooks} -> center: {mating_deltas}")
    #display(board)
    #print('------------------------------------------------------')
    #print()

    df.loc[line, 'Mating_Color'] = f"{'white' if mating_color else 'black'}"
    df.loc[line, 'square_0_0'] = chess.square(file_index=0, rank_index=0)

    if valid_pattern is True:
      if valid_pattern_bishop:
        df.loc[line, "Queen_Bishop"] = True

      if valid_pattern_pawn:
        df.loc[line, "Queen_Pawn"] = True
    else:
      line = line + 1
      continue

    # Print the board position #################################################
    # print(board)
    # display(board)
    # print('DEBUGGING')

    line = line + 1
    matched_lines = matched_lines + 1

    # Break after x matches (for debugging only - when developing the script)
    #if (matched_lines > 50000):
    #  break
```

```
print(f"{matched_lines} Damianos mate patterns found.")


# Output the first 100 rows of the DataFrame
filtered_df = df[(df['Queen_Pawn'] == True) | (df['Queen_Bishop'] == True)] # Filter where 'Rook_Knight' is True
result = filtered_df.head(100)  # Get the first 100 rows

print(result)  # Display the result
```
